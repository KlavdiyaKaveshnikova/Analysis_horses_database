**Сделаю то же самое через квартили.**

```python
def outliers(column):
    q1 = cut_data[column].quantile(0.25)
    q3 = cut_data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = cut_data[cut_data[column].between(lower_bound, upper_bound, inclusive=True)].sort_values(column)
    return remove_outliers
```
